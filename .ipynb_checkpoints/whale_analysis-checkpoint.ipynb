# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp_tsx_history.csv`: Contains historical closing prices of the S&P TSX 60 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data.

```python
# Reading whale returns
whale_ret_csv = Path("Resources/whale_returns.csv")
whale_ret_df = pd.read_csv(whale_ret_csv)

# Convert Dates to DateTimeIndex
whale_ret_df = pd.read_csv(whale_ret_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
whale_ret_df.head()
```

```python
# Count nulls - counting the missing values in each row
whale_ret_df.isnull().sum(axis=1)
```

```python
# Drop nulls
whale_ret_df = whale_ret_df.dropna()
whale_ret_df.head()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data.

```python
# Reading algorithmic returns
algo_ret_csv = Path("Resources/algo_returns.csv")
algo_ret_df = pd.read_csv(algo_ret_csv)

algo_ret_df = pd.read_csv(algo_ret_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
algo_ret_df.head()
```

```python
# Count nulls
algo_ret_df.isnull().mean()*100
algo_ret_df.isnull().sum()
```

```python
# Drop nulls
algo_ret_df = algo_ret_df.dropna()
algo_ret_df.head()
```

## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P TSX 60 Closing Prices
sptsx60_ret_csv = Path("Resources/sp_tsx_history.csv")
sptsx60_ret_df = pd.read_csv(sptsx60_ret_csv)

sptsx60_ret_df = pd.read_csv(sptsx60_ret_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
sptsx60_ret_df.head()
```

```python
# Check Data Types
sptsx60_ret_df.dtypes
```

```python
# Remove dollar sign. Got an error fixing datat types without removing a non-numeric data i.e '$' sign  
sptsx60_ret_df['Close'] = sptsx60_ret_df['Close'].replace('\$|,', '', regex=True)
sptsx60_ret_df['Close']

#sptsx60_ret_df = sptsx60_ret_df.sort_values("Date", ascending=True)
sptsx60_ret_df['Close'] = sptsx60_ret_df['Close'].astype('float')
sptsx60_ret_df['Close'].dtype
```

```python
# Calculate Daily Returns
sptsx60_ret_df.isnull().sum()
sptsx60_ret_df = sptsx60_ret_df.pct_change()
sptsx60_ret_df.head()
```

```python
# Drop nulls
sptsx60_ret_df.pct_change
sptsx60_ret_df.dropna(inplace=True)
sptsx60_ret_df.head()
```

```python
# Rename `Close` Column to be specific to this portfolio.
sptsx60_ret_df = sptsx60_ret_df.rename(columns={"Close":"S&PTSX"})
sptsx60_ret_df.head()
```

## Combine Whale, Algorithmic, and S&P TSX 60 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
combined_portfolio_df = pd.concat([whale_ret_df, algo_ret_df, sptsx60_ret_df], axis='columns', join='inner')
combined_portfolio_df.dropna(inplace=True)
combined_portfolio_df.head()
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for all portfolios using a 21-day window.
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P TSX 60.
3. Choose one portfolio, then calculate and plot the 60-day rolling beta for it and the S&P TSX 60.

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window

# Plot the rolling standard deviation
combined_portfolio_df.rolling(window=21).std().plot(title="21 Day Rolling Standard Deviation", figsize=(20,10))
```

### Calculate and plot the correlation

```python
# Calculate the correlation
correlation = combined_portfolio_df.corr()
# Display de correlation matrix
correlation
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 60 TSX

```python
# Calculate covariance of a single portfolio
algo2_cov = combined_portfolio_df['Algo 2'].rolling(window=60).cov(combined_portfolio_df['S&PTSX'])
algo2_cov
# Calculate variance of S&P TSX
sptsx_var = combined_portfolio_df['S&PTSX'].rolling(window=60).var()
sptsx_var
# Computing beta

# Plot beta trend
algo2_beta = algo2_cov / sptsx_var
algo2_beta_plot = algo2_beta.plot(title="Algo 2 Beta", figsize=(20,10))
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half life for each portfolio, using standard deviation (`std`) as the metric of interest.

```python
# Use `ewm` to calculate the rolling window
rolling_window_ewm = combined_portfolio_df.ewm(halflife=21, adjust=True).std().plot(title="Exponentialy Weighted Average", figsize=(20,10))
rolling_window_ewm
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratios = (combined_portfolio_df.mean() * 252) / (combined_portfolio_df.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Reading data from 1st stock
l_history_csv = Path("Resources/l_historical.csv")
l_history_df = pd.read_csv(l_history_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
l_history_df.head()
```

```python
# Reading data from 2nd stock
otex_history_csv = Path("Resources/otex_historical.csv")
otex_history_df = pd.read_csv(otex_history_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
otex_history_df.head()
```

```python
# Reading data from 3rd stock
shop_history_csv = Path("Resources/shop_historical.csv")
shop_history_df = pd.read_csv(shop_history_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
shop_history_df.head()
```

```python
# Combine all stocks in a single DataFrame
combined_histport_df = pd.concat([l_history_df, otex_history_df, shop_history_df], axis="columns", join="inner")
combined_histport_df
```

```python
# Reset Date index
combined_histport_df.sort_index(inplace=True)
combined_histport_df.drop(columns=['Symbol'], inplace=True)
```

```python
# Reorganize portfolio data by having a column per symbol
combined_histport_df.columns = ['L', 'OTEX', 'SHOP']
combined_histport_df.head()
```

```python
# Calculate daily returns
combined_daily_returns = combined_histport_df.pct_change()
# Drop NAs
combined_daily_returns.dropna(inplace=True)
# Display sample data
combined_daily_returns.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/3, 1/3, 1/3]

# Calculate portfolio return
combined_weighted_returns = combined_daily_returns.dot(weights)
combined_weighted_returns.head()
# Display sample data
combined_weighted_returns = combined_weighted_returns.rename("Custom Portfolio")
combined_weighted_returns.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
# need to initialize new dataframe - without this an error name not defined will occur
combined_orig_portfolio_df = pd.concat([whale_ret_df, algo_ret_df, sptsx60_ret_df, combined_weighted_returns], axis='columns', join='inner')
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
combined_orig_portfolio_df.dropna(inplace = True)
combined_orig_portfolio_df.head()
```

## Re-run the risk analysis with your portfolio to see how it compares to the others

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized `std`
daily_std=combined_orig_portfolio_df.std()
annualised_std = daily_std * np.sqrt(252)
annualised_std = annualised_std.sort_values(ascending = False)
annualised_std.head(10)
```

### Calculate and plot rolling `std` with 21-day window

```python
# Calculate rolling standard deviation

# Plot rolling standard deviation
combined_orig_portfolio_df.rolling(window=21).std().plot(figsize=(20,10))
```

### Calculate and plot the correlation

```python
# Calculate and plot the correlation
combined_port_corr_df = combined_orig_portfolio_df.corr()
combined_port_corr_df
```

### Calculate and Plot the 60-day Rolling Beta for Your Portfolio compared to the S&P 60 TSX

```python
# Calculate and plot Beta

# calculate Beta
custom_covariance = combined_orig_portfolio_df['Custom Portfolio'].rolling(window=60).cov(combined_orig_portfolio_df['S&PTSX'])
custom_covariance

# Calculate variance of all daily returns vs. S&P 500
sp_variance = combined_orig_portfolio_df['S&PTSX'].rolling(window=60).var()
sp_variance

# Calculate beta of all daily returns 
custom_beta = custom_covariance / sp_variance
custom_beta_plot = custom_beta.plot(title="60-day Rolling Beta", figsize=(15, 8))
```

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Calculate Annualized Sharpe Ratios
combined_sharpe_ratios_df = (combined_orig_portfolio_df.mean() * 252) / (combined_orig_portfolio_df.std() * np.sqrt(252))
combined_sharpe_ratios_df.sort_values(ascending=False)
```

```python
# Visualize the sharpe ratios as a bar plot
combined_sharpe_ratios_df.plot(kind="bar", title="Sharpe Ratios")
```
